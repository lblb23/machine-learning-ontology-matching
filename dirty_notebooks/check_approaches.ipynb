```python
def t_ocs(x,y):
    n = len(x) + 1
    m = len(y) + 1
    diag = n*m + m
    d = [0] * (diag+n+1)
    for i in range(1,n):
        diag = diag + 1
        
        for j in range(1,m):
            if d[(j-1)*n+i] < d[j*n+i-1]:
                d[j*n + i] = d[(j-1)*n+i]
            else:
                d[j*n + i] = d[j*n+i-1]
            
                if (x[i-1] == y[j-1]):
                    d[diag-j] = d[diag-j] + 1

                    if (d[j*n+i] < d[j*n+i-n-1] + d[diag-j]):
                        d[j*n+i] = d[j*n+i-n-1] + d[diag-j]
                else:
                    t = d[n*m-1]
                    return t
```
